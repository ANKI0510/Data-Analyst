```python
# 5.	Use a boxplot on the rating column. Use User Defined Function or Lambda function or
#Apply function to extract the data that comes before /. Kindly exclude /.
df['rate'] = df['rate'].replace('NEW',np.nan)
df['rate'] = df['rate'].replace('-',np.nan)
df['rate'] = df['rate'].replace('nan',np.nan)
df['rate'] = df['rate'].replace('NaN',np.nan)

```

```python
X = df
X.rate = X.rate.astype(str)
X.rate = X.rate.apply(lambda x:x.replace('/5',''))
X.rate = X.rate.apply(lambda x:float(x))
X.head()
```

```python
# 8.	Find the distribution of the votes and Approx_cost using a user-defined function and for a loop.
def plot_distribution(column_name):
  plt.figure(figsize=(10,5))
  sns.histplot(df[column_name],kde=True)
  plt.xlabel(column_name)
  plt.xticks(rotation=90)
  plt.ylabel('Frequency')
  plt.show()
column_to_analyze = ['votes','approx_cost(for two people)']
for i in column_to_analyze:
  plot_distribution(i)
```

```python
# 9.	Which are the most common restaurant type in Banglore?
df['rest_type'].value_counts()
```

```python
# 10.Is there any difference b/w the votes of restaurants accepting and not accepting online orders?

accepting_online_order =df[df['online_order']=='Yes']
not_accepting_online_order =df[df['online_order']=='No']

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.barplot(x='online_order',y='votes',data=accepting_online_order)
plt.title('Votes for restaurant accepting online order')

plt.subplot(1,2,2)
sns.barplot(x='online_order',y='votes',data=not_accepting_online_order)
plt.title('Votes for restaurant not accepting online order')
plt.tight_layout()
plt.show()
```

```python
#14.	Which are the most popular casual dining restaurant chains, Make use of any plot related to this question?
df[df['rest_type']=='Casual Dining'].sort_values('votes',ascending=False).head(10)
```

```python

desired_location = input('Enter a location')
casual_dining_rest = df[df['location']==desired_location & (df['rest_type']=='Casual Dining')]
popular_casual_dining = casual_dining_rest.groupby('name')['votes'].sum().reset_index()
popular_casual_dining = popular_casual_dining.sort_values(by='votes',ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x='votes',y='name',data=popular_casual_dining.head(10))
plt.xlabel('Restaurant Name')
plt.ylabel('Total Votes')
plt.title(f'Most popular casual dining in{desired_location}')
plt.tight_layout
plt.show()
```
