1. A survey was conducted and it was found that the people spend their 300 minutes
in a day surfing on online shopping sites on average and the corresponding standard deviation is 127 minutes.
Assume that the time spent on surfing follows a normal distribution.
Calculate the following probabilities:
a. What is the probability that the users are spending less than or equal to 100 minutes per day?
b. What is the probability that people are spending more than 400 minutes on online shopping sites per day?
c. What is the probability that people are spending time between 250 minutes and 350 minutes per day?

```python
#a. What is the probability that the users are spending less than or equal to 100 minutes per day?

# while applying z score mean would be 0 and standard deviation(S.D) = 1
# method 1
mean = 300
sd = 127
x = 100
z = (x-mean)/sd

stats.norm.cdf(z,0,1)
```

```python
# Method 2
stats.norm.cdf(x,mean,sd)
```

```python
# Method 3 python bydefault understood the value of z that mean = 0 and s.d = 1
stats.norm.cdf(z)
```

```python
# b. What is the probability that people are spending more than 400 minutes on online shopping sites per day?
mean = 300
sd = 127
x = 400
z = (x-mean)/sd
stats.norm.sf(x,mean,sd)
```

```python
1-stats.norm.cdf(z)
```

```python
# c. What is the probability that people are spending time between 250 minutes and 350 minutes per day?
mean = 300
sd = 127
x1 = 250
x2 = 350
z1 = (x1-mean)/sd
z2 = (x2-mean)/sd
stats.norm.sf(x1,mean,sd)
```

```python
stats.norm.cdf(z2) - stats.norm.cdf(z1)
```

```python
stats.norm.sf(z1) - stats.norm.sf(z2)
```

# Central Limit Theory


```python
# 3> Create a sampling distribution from a right skewed population of size 100000 and a skewness of +20
# Create a sampling distribution from the above population with replacement -with sample size = 30 and number of samples = 1000

population_array = stats.skewnorm.rvs(a = +20,size = 100000,random_state = 0)
sns.distplot(population_array)
plt.title('Population Distribution')
```

```python
# Create a sampling distribution from the above population with replacement -with sample size = 30 and number of samples = 1000
sample_size = 30
n_samples = 1000
sample_means = []
for i in range(n_samples):
  sample = np.random.choice(population_array,size=sample_size,replace = True)
  sample_mean = np.mean(sample)
  sample_means.append(sample_mean)

sns.distplot(sample_means)
plt.title('Sampling Distribution')
plt.show()

```

```python
## Comparing Population mean and sampling distribution means

round(np.mean(population_array),2)
```

```python
round(np.mean(sample_means),4)
```

```python
## Compare standard dev

round(np.std(population_array)/np.sqrt(sample_size),2)
```

```python
round(np.std(sample_means,ddof=1),2) # Delta Degree of freedom
```

# Normality Test

```python
# The temperature (in degree Celsius) at Manhattan is recorded for 45 winter days. A student from the Geography department wants to test the claim proposed by the weather department.
# Check whether the given data satisfies the condition of normality with 99% confidence?
# Use the data given below:

temp_winter = [2.51, 2.6, 1.8, 2.61, 3.9, 1.7, 1.9, 3.1, 1.58, 2.85, 2.9, 1.25, 1.3, 2.4, 3.22, 2.56, 1.32, 1.2, 1.26, 3.2,
               2.7, 3.5, 2.8, 1.5, 3.8, 1.84, 3.31, 1.24, 1.75, 1.41, 1.57, 1.4, 1.49, 2.26, 2.5, 2.52, 3.42, 1.36, 2.35, 2.47, 2.28, 2.22, 2.66, 3.43, 1.98]

stats.shapiro(temp_winter)     ### To Check Normality
```

```python
## p- value > .01 (1% level of significance) -------> Data is normal
## We can perform hypotheisis test
```

```python
 # The average calories in a slice bread of the brand 'Alphas' are 82 with a standard deviation of 15.
 # An experiment is conducted to test the claim of the dietitians that the calories in a slice of bread are not as per the manufacturer's specification.
 # A sample of 40 slices of bread is taken and the mean calories recorded are 95.
 # Test the claim of dietitians with  value (significance level) as 0.05. (assume the normality of the data).
## H0 : mu = 82
## H1 : mu !=82

mu = 82
n = 40
x_bar = 95
sigma = 15

z_stat = (x_bar-mu)/(sigma/n**.5)
z_stat
```

```python
z_critical = stats.norm.isf(.025)
z_critical
```

```python
z_stat > z_critical ## thus reject null hypothesis and accept alternative
# z_ critical are -1.95 and + 1.95
# if z_ stat > 1.95 or < -1.95 then we reject H0   and  accept it
# Therefore calories are not as per specifications
```

```python
p_value = 2* stats.norm.sf(z_stat) ## Two tail test
p_value
```

```python
#The internet provider claims that their router gives a speed of 25 Mbps with a standard deviation of 4.8 Mbps.
#The computer engineer wants to check whether the speed is different than what the provider claims. For this study,
#the engineer collects the speed of 40 routers
#and found that the average speed is 23.2 Mbps for those 40 routers. Test the claim of engineer with 95% confidence.

# H0 : mu = 25
# H1 : mu != 25

mu = 25
n = 40
x_bar = 23.2
sigma = 4.8

z_stat = (x_bar-mu)/(sigma/n**.5)
z_stat
```

```python
# Thus z_stat is negative so use ppf and cdf
z_critical = stats.norm.ppf(.025)
z_critical
```

```python
p_value = 2* stats.norm.cdf(z_stat) ## Two tail test
p_value
```

```python
# since p_value < 0.05 , thus reject null
# We can  conclude that the average speed is different from 25.
```
